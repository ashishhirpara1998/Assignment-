## Q1. What is Flask Framework? What are the advantages of Flask Framework?


Flask is a lightweight and flexible Python web framework that provides tools, libraries, and utilities for building web applications. It is known for its simplicity and minimalistic design, making it easy to get started with and suitable for small to medium-sized projects. Flask is often referred to as a "micro" framework because it does not come with built-in features like form validation, database abstraction, or authentication, but it allows developers to choose and integrate third-party extensions based on their specific needs.

Advantages of Flask Framework:

1. Lightweight and Minimalistic: Flask has a simple and minimalistic design, which means it has a small codebase and does not impose unnecessary dependencies. This makes it easy to understand and maintain, especially for developers who prefer simplicity over complexity.

2. Flexibility: Flask provides developers with the flexibility to choose the components and libraries they need for their project. It follows a "you only pay for what you use" philosophy, allowing developers to add only the features they require without being burdened by unnecessary functionality.

3. Extensibility: Flask is highly extensible through its ecosystem of extensions. These extensions cover a wide range of functionalities such as form validation, database integration, authentication, and more. Developers can easily integrate these extensions into their Flask applications to add additional features without reinventing the wheel.

4. Easy to Get Started: Flask has a straightforward and intuitive API, making it easy for developers to get started with building web applications. Its documentation is also comprehensive and beginner-friendly, providing clear guidance on how to use its features effectively.

5. Built-in Development Server: Flask comes with a built-in development server, which makes it easy to develop and test applications locally without the need for additional setup. This development server is convenient for rapid prototyping and debugging during the development process.

6. Pythonic: Flask follows the principles of Python, making it easy for Python developers to work with. It leverages Python's syntax and conventions, which results in clean and readable code.

7. Large Community and Ecosystem: Flask has a large and active community of developers who contribute to its ecosystem by creating extensions, tutorials, and resources. This vibrant community ensures that developers have access to a wealth of resources and support when building Flask applications.

## Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to specific functions or view handlers within the application. This mapping is defined using the @app.route() decorator in Flask.

We use app routes in Flask for several reasons:

1. URL Mapping: App routes allow us to define the URL structure of our web application and map specific URLs to corresponding functions or view handlers.

2. Organized Code: By using app routes, we can organize our codebase effectively. Each route and its associated functionality are defined within a single function, making it easy to manage and maintain the code.

3. RESTful Routing: Flask supports the principles of RESTful routing, where different HTTP methods (GET, POST, PUT, DELETE, etc.) are used to perform different actions on resources. App routes help in defining these routes for various HTTP methods.

4. Modular Development: App routes enable modular development, allowing different parts of the application to be developed and maintained independently. Each route can be implemented and tested separately, enhancing code reusability and maintainability.

5. Dynamic Routing: Flask supports dynamic routing, where route parameters can be included in the URL pattern. This allows for more flexible and dynamic URL handling, such as passing variables through the URL.

## Q4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    '''

if __name__ == '__main__':
    app.run(debug=True)
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.


In Flask, the **url_for()** function is used for URL building. It generates a URL to a specific function within the application. This function is particularly useful because it allows you to avoid hardcoding URLs in your templates or view functions, making your application more flexible and easier to maintain, especially if you need to change URLs later.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    # Generating URL for the 'about' function
    about_url = url_for('about')
    return f'The URL for the about page is: {about_url}'

@app.route('/about')
def about():
    return 'This is the about page'

if __name__ == '__main__':
    app.run(debug=True)
```
